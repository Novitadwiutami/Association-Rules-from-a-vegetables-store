```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import OnehotTransactions
```

```python
dataset = [['Broccoli', 'Green Peppers', 'Corn','Tomatoes','Beans'],
           ['Asparagus', 'Squash', 'Corn','Tomatoes'],
           ['Corn', 'Tomatoes', 'Beans', 'Green Peppers','Broccoli'],
           [ 'Corn', 'Tomatoes', 'Beans','Broccoli'],
           ['Broccoli','Green Peppers','Beans','Asparagus'],
          ['Green Peppers','Asparagus','Beans','Tomatoes'],
          ['Tomatoes','Corn','Beans','Asparagus','Green Peppers','Broccoli']]
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
print(df)
```

```python
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
print (frequent_itemsets) 
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
```

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
```

```python
rules[ (rules['antecedent_len'] >= 2) &
       (rules['confidence'] > 0.75) &
       (rules['lift'] > 1.2) ]
```

```python
rules[rules['antecedents'] == {'Beans', 'Corn'}]
```

```python
rules[rules['antecedents'] == {'Asparagus','Tomatoes'}]
```
